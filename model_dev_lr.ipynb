```python
# Imports
import pandas as pd
import pickle
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.pipeline import Pipeline
```

```python
# Load preprocessed data splits for Linear Regression, SVR, Neural Networks (for df_1 and df_2)
with open('logistic_svm_nn_splits_df1.pkl', 'rb') as f:
    X_train_1, X_test_1, y_train_1, y_test_1 = pickle.load(f)

with open('logistic_svm_nn_splits_df2.pkl', 'rb') as f:
    X_train_2, X_test_2, y_train_2, y_test_2 = pickle.load(f)
```

```python
# Creating pipeline for Linear Regression
model_pipeline = Pipeline(steps=[('regressor', LinearRegression())])
```

```python
# Training model on df_1
model_pipeline.fit(X_train_1, y_train_1)
```

```python
# Making predictions on df_1
y_pred_1 = model_pipeline.predict(X_test_1)
```

```python
# Evaluating model on df_1
mse_1 = mean_squared_error(y_test_1, y_pred_1)
r2_1 = r2_score(y_test_1, y_pred_1)
```

```python
# Displaying evaluation metrics for df_1
print("df_1 Linear Regression Test Results")
print(f'MSE: {mse_1}, R2: {r2_1}')
print("="*60)
```

```python
# Training model on df_2
model_pipeline.fit(X_train_2, y_train_2)
```

```python
# Making predictions on df_2
y_pred_2 = model_pipeline.predict(X_test_2)
```

```python
# Evaluating model on df_2
mse_2 = mean_squared_error(y_test_2, y_pred_2)
r2_2 = r2_score(y_test_2, y_pred_2)
```

```python
# Printing evaluation metrics for df_2
print("df_2 Linear Regression Test Results")
print(f'MSE: {mse_2}, R2: {r2_2}')
print("="*60)
```
